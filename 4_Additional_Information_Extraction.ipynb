Document and symbol extractions

```python
import re
# Define the folder containing text documents
input_folder = "/content/drive/MyDrive/Stanford/TEXT 1 page"
# Function to extract dates and symbols from text and return them as comma-separated strings
def extract_dates_and_symbols(text):
    date_formats = [
        r'\d{4}'                      # YYYY
        r'\[a-zA-Z]+ \d{4}'
        r'\d{1,2}- \[a-zA-Z]+ \d{4}',     # MM- Month YYYY
        r'\d{1,2}/\d{1,2}/\d{4}',     # MM/DD/YYYY
        r'\d{1,2}-\d{1,2}-\d{4}',     # MM-DD-YYYY
        r'\d{4}/\d{1,2}/\d{1,2}',     # YYYY/MM/DD
        r'\d{4}-\d{1,2}-\d{1,2}',     # YYYY-MM-DD
        r'\d{1,2}/\d{1,2}/\d{2}',     # MM/DD/YY
        r'\d{1,2}-\d{1,2}-\d{2}',     # MM-DD-YY
        r'\d{1,2}\.\d{1,2}\.\d{4}',   # dd.mm.yyyy
        r'\d{1,2} [a-zA-Z]+ \d{4}',   # dd month yyyy
        r'\d{1}\.\d{1,2}\.\d{4}',     # d.mm.yyyy
        r'\d{1}\.\d{1}\.\d{4}',       # d.m.yyyy
        r'\d{1}\.\d{1}\.\d{2}',       # d.m.yy
        r'[a-zA-Z]+ \d{4}',           # month yyyy
        r'\d{1,2}- [a-zA-Z]+ \d{4}',  # dd- month yyyy
    ]

    symbol_formats = [r'[a-zA-Z]+\.\d+/\d+', r'[a-zA-Z]+\/\d+', r'[a-zA-Z]+\(\d+\)\d+', r'd+/\d+', r'd+\-/d+',
                      r'[a-zA-Z]+\. \d+/\d+',  r'[a-zA-Z]+\d+\-\d+/\d+',r'd+\.\d+/\d+']

    dates = []
    symbols = []

    for symbol_format in symbol_formats:
        symbols += re.findall(symbol_format, text)

    for date_format in date_formats:
        dates += re.findall(date_format, text)

    # Combine dates and symbols into comma-separated strings
    dates_str = ', '.join(dates)
    symbols_str = ', '.join(symbols)

    return dates_str, symbols_str

# Create a CSV file to store the data
csv_file_path = os.path.join(input_folder, "document_data.csv")
with open(csv_file_path, "w", newline="", encoding="utf-8") as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(["Document Name", "Dates", "Symbols", "Full Text"])

    # Loop through all files in the input folder
    for filename in os.listdir(input_folder):
        if filename.endswith(".txt"):  # Assuming the documents are text files
            # Build the full path to the input file
            input_file_path = os.path.join(input_folder, filename)

            # Open the input file for reading
            with open(input_file_path, "r", encoding="utf-8") as input_file:
                # Read the content of the input file
                text_content = input_file.read()

                # Extract dates and symbols
                dates_str, symbols_str = extract_dates_and_symbols(text_content)

                # Write data to the CSV file
                csv_writer.writerow([filename, dates_str, symbols_str, text_content])

print(f"CSV file '{csv_file_path}' created.")
```

Title extraction from Text files

```python
import os
import re
import csv

# Define the folder containing text files
text_folder = '/content/drive/MyDrive/Stanford/TEXT 1 page'

# Initialize a list to store extracted titles
extracted_titles = []

# Initialize a CSV file to store the results
csv_file_path = '/content/drive/MyDrive/Stanford/TEXT 1 page/extracted_titles.csv'
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(["Filename", "Title"])  # Write header row

    # Iterate through text files in the directory
    for filename in os.listdir(text_folder):
        if filename.endswith('.txt'):
            text_file_path = os.path.join(text_folder, filename)

            # Read the content of the text file
            with open(text_file_path, 'r', encoding='utf-8') as text_file:
                lines = text_file.readlines()

                # Initialize variables to store date and title
                date = None
                title = None

                # Search for dates in the text content
                date_formats = [
                    r'\d{1,2}/\d{1,2}/\d{4}',     # MM/DD/YYYY
                    r'\d{1,2}-\d{1,2}-\d{4}',     # MM-DD-YYYY
                    r'\d{4}/\d{1,2}/\d{1,2}',     # YYYY/MM/DD
                    r'\d{4}-\d{1,2}-\d{1,2}',     # YYYY-MM-DD
                    r'\d{1,2}/\d{1,2}/\d{2}',     # MM/DD/YY
                    r'\d{1,2}-\d{1,2}-\d{2}',     # MM-DD-YY
                    r'\d{1,2}\.\d{1,2}\.\d{4}',   # dd.mm.yyyy
                    r'\d{1,2} [a-zA-Z]+ \d{4}',   # dd month yyyy
                ]

                # Flags to indicate whether to search for the next title
                search_next_title = False

                for line in lines:
                    if search_next_title:
                        # Check if the line contains letters (indicating a title)
                        if any(char.isalpha() for char in line):
                            title = line.strip()
                            search_next_title = False
                            break  # Stop searching for titles once a title is found

                    # Search for dates in the line
                    for date_format in date_formats:
                        dates = re.findall(date_format, line)
                        if dates:
                            date = dates[0]
                            search_next_title = True
                            break  # Stop searching for dates once a date is found

                # If a title is found, add it to the list and CSV file
                if title:
                    extracted_titles.append((filename, title))
                    csv_writer.writerow([filename, title])

# Print the extracted titles on the screen
for filename, title in extracted_titles:
    print(f"Filename: {filename}, Title: {title}")

print(f"Extraction completed. Results saved to {csv_file_path}.")
```

```python
import os
import csv

# Set the directory containing text files
text_directory = '/content/drive/MyDrive/Stanford/TEXT 1 page'

# Initialize variables to store titles and results
titles = []
results = []

# Function to check if a string contains all capital letters
def is_all_caps(s):
    return s.isupper()

# Function to check if a string matches excluded expressions
def is_excluded(s):
    excluded_patterns = [
        "RESTRICTED",
        "CONFIDENTIAL",
        "GENERAL AGREEMENT",
        "TARIFFS AND TRADE",
        "/",
        "Special Distribution"
        "Limited Distribution"
    ]
    return any(pattern in s for pattern in excluded_patterns) or any(char.isdigit() for char in s)

# Function to find titles in a text file
def find_titles_in_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    current_title = ""
    current_filename = os.path.basename(file_path)

    for i in range(len(lines)):
        line = lines[i].strip()

        if is_all_caps(line) and not is_excluded(line):
            # Check the line above
            if i > 0 and not lines[i - 1].strip():
                current_title = lines[i - 1].strip() + ' ' + line
            else:
                current_title = line
            titles.append(current_title)
            results.append([current_filename, current_title])
            current_title = ""

# Iterate through text files in the directory
for filename in os.listdir(text_directory):
    if filename.endswith('.txt'):
        file_path = os.path.join(text_directory, filename)
        find_titles_in_file(file_path)

# Print the filename and title
for filename, title in results:
    print(f"Filename: {filename}, Title: {title}")

# Save the results to a CSV file
csv_file_path = '/content/drive/MyDrive/Stanford/TEXT 1 page/output.csv'  # Set the desired output CSV file path
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(["Filename", "Title"])
    csv_writer.writerows(results)

print("Results saved to CSV file:", csv_file_path)
```

```python
import PyPDF2
import re
import os
import csv

# Function to extract titles from a PDF file
def extract_title_from_pdf(pdf_file_path):
    # Initialize a PDF file reader
    pdf_reader = PyPDF2.PdfReader(open(pdf_file_path, 'rb'))

    # Initialize variables to track the title
    title = ""
    is_title_line = False

    # Loop through each page and extract text
    for page_num in range(len(pdf_reader.pages)):
        page = pdf_reader.pages[page_num]
        page_text = page.extract_text()

        # Split the text into lines
        lines = page_text.split('\n')

        # Check each line for the title pattern
        for line in lines:
            # Check if the line is in all capital letters or if it's underlined
            if line.isupper() or re.match(r'^[-=_]{4,}$', line):
                is_title_line = True
                title += line.strip() + " "
            else:
                # If the previous line was a title line, break
                if is_title_line:
                    break

    # Clean up the extracted title
    title = title.strip()
    return title

# Directory containing PDF files
pdf_directory = '/content/drive/MyDrive/Stanford/PDF 1 page'

# CSV file to store the results
csv_file_path = '/content/drive/MyDrive/Stanford/PDF 1 page/pdf_titles_output.csv'

# Initialize a list to store results
results = []

# Loop through all PDF files in the directory
for filename in os.listdir(pdf_directory):
    if filename.endswith('.pdf'):
        pdf_file_path = os.path.join(pdf_directory, filename)
        title = extract_title_from_pdf(pdf_file_path)
        results.append([filename, title])
        print(f"File: {filename}, Title: {title}")

# Save the results to a CSV file
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(["Filename", "Title"])
    csv_writer.writerows(results)

print("Results saved to CSV file.")
```

```python
import os
import csv
from tika import parser

# Function to extract titles from a PDF file using Apache Tika
def extract_title_from_pdf(pdf_file_path):
    try:
        # Parse the PDF file using Apache Tika
        parsed_pdf = parser.from_file(pdf_file_path)

        # Get the content text from the parsed PDF
        pdf_text = parsed_pdf['content']

        # Split the text into lines
        lines = pdf_text.split('\n')

        # Initialize the title variable
        title = ""

        # Regular expressions to check for unwanted patterns
        unwanted_patterns = [
            r'RESTRICTED',
            r'CONFIDENTIAL',
            r'GENERAL AGREEMENT',
            r'TARIFFS AND TRADE',
            r'/',
            r'\d+',
            r'ON TARIFFS AND',
            r'TRADE'
            r"GATT/"
            r"MORE"
        ]


        # Loop through the lines to find the title
        for line in lines:
            # Check if the line contains any unwanted patterns
            if any(re.search(pattern, line) for pattern in unwanted_patterns):
                continue

            # Check if the line contains capital letters or is underlined
            if line.strip().isupper() or line.strip().startswith("_____"):
                title = line.strip()

                # Check if the title contains any unwanted patterns
                if any(re.search(pattern, title) for pattern in unwanted_patterns):
                    # Search for an alternative title in the next line
                    for next_line in lines[lines.index(line) + 1:]:
                        if next_line.strip().isupper() or next_line.strip().startswith("_____"):
                            title = next_line.strip()
                            break

                break

        return title
    except Exception as e:
        print(f"Error processing {pdf_file_path}: {e}")
        return ""

# Directory containing PDF files
pdf_directory = '/content/drive/MyDrive/Stanford/PDF 1 page'

# CSV file to store the results
csv_file_path = '/content/drive/MyDrive/Stanford/PDF 1 page/pdf_titles.csv'

# Initialize a list to store results
results = []

# Loop through all PDF files in the directory
for filename in os.listdir(pdf_directory):
    if filename.endswith('.pdf'):
        pdf_file_path = os.path.join(pdf_directory, filename)
        title = extract_title_from_pdf(pdf_file_path)
        results.append([filename, title])
        print(f"File: {filename}, Title: {title}")

# Save the results to a CSV file
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(["Filename", "Title"])
    csv_writer.writerows(results)

print("Results saved to CSV file.")
```
